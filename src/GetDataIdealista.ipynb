## Obtener datos de Idealista

```python
idealista = idealistatools.Idealista(debug = True)
center = '40.44,-3.71'
distance = '20000'
```

```python
my_api_key = ""
my_api_secret = ""
idealista.generate_token(api_key=my_api_key, api_secret=my_api_secret)
```

```python
totalPages = 1382
pagIni = 1
for pagina in range(pagIni, totalPages):
    if (pagina - 1) % 20 == 0:
        print(str(pagina))
    result = idealista.search(numPage=str(pagina),center=center,distance=distance)
    summary = idealista.summary_result(result)
    idealista.elementlist_tojson(result, 'idealista_' + str(pagina) + ".json")
    time.sleep(10)
```
